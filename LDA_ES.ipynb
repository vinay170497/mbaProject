```python
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from gensim.models import LdaModel
from gensim.corpora import Dictionary
nltk.download('punkt')
nltk.download('stopwords')
import pandas as pd
```

```python
def preprocess(textstring):
   stops =  set(stopwords.words('english'))
   custom_stop_words = {'film','will','whether','like','also','much','one','get','go','well','day','bad','great','nice','first','would','good','use',
                       'alright','guy','really','thing','guy','lie','help','hard','need','rd','mujhe','n'}
   stops.update(custom_stop_words)
   tokens = word_tokenize(textstring)
   return [token.lower() for token in tokens if token.isalpha() and token not in stops]
```

```python
filepath_Tablet = r"D:\PGPM\Empirical Study\Datasets\Amazon_Tablet_Reviews.xlsx"
filepath_Laptop = r"D:\PGPM\Empirical Study\Datasets\Amazon_Laptop_Reviews.xlsx"
filepath_Mobile = r"D:\PGPM\Empirical Study\Datasets\Amazon_Mobile_Reviews.xlsx"
df1=pd.read_excel(filepath_Tablet)
df2=pd.read_excel(filepath_Laptop)
df3=pd.read_excel(filepath_Mobile)

print("Tablet Dataframe Head.")
print(df1.info())
print("****************************************")
print("Laptop Dataframe Head.")
print(df2.info())
print("****************************************")
print("Mobile Dataframe Head.")
print(df3.info())
```

```python
#Topic Modelling for Tablets.
total_reviews = " ".join(str(text) for text in df1.new_reviews)
#print(type(total_reviews))
preprocess(total_reviews)
```

```python
cleaned_tokens=[]
sent_tokens=sent_tokenize(total_reviews)
for line in sent_tokens:
    cleaned_tokens.append(preprocess(line))

print(cleaned_tokens)
```

```python
my_dictionary = Dictionary(cleaned_tokens)
print(len(my_dictionary))
```

```python
my_corpus = [my_dictionary.doc2bow(c) for c in cleaned_tokens]

#This is the number of documents
print(len(my_corpus))
```

```python
#Corpus has the bag of words
word_id, freq = my_corpus[0][4]
#the word is in the dictionary
word = my_dictionary[word_id]
print(word)

my_id2word = my_dictionary.id2token
print(my_id2word)
```

```python
my_number_of_topics=3
myLDAmodel = LdaModel(corpus=my_corpus, id2word=my_id2word,iterations=10000, num_topics=my_number_of_topics)

for idx in range(my_number_of_topics):
    print("Topic #%s:" % idx, myLDAmodel.print_topic(idx, 5))

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
vis = gensimvis.prepare(topic_model=myLDAmodel, 
                              corpus=my_corpus, 
                              dictionary=my_dictionary)
pyLDAvis.display(vis)
```
